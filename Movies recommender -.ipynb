```python
import numpy as np
import pandas as pd
import ast
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.columns
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3,max_features=5000,strip_accents = 'unicode',analyzer='word',token_pattern=r'\w{1,}',stop_words='english')
```

```python
tfv_matrix = tfv.fit_transform(new['tags'])
```

```python
tfv_matrix.shape
```

```python
tfv.get_feature_names()
```

```python
tfv_similarity = cosine_similarity(tfv_matrix)
```

```python
tfv_similarity
```

```python
def recommend_tfv(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(tfv_similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend_tfv('Avatar')
```
